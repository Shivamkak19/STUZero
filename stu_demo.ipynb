# STU Implementation for Learning LDSs

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

torch.manual_seed(0)
np.random.seed(0)

# =============================================================================
# Utility Functions
# =============================================================================

def nearest_power_of_two(x: int, round_up: bool = False) -> int:
    """Find the nearest power of 2 to x."""
    if not round_up:
        return 1 << math.floor(math.log2(x))
    else:
        return 1 << math.ceil(math.log2(x))

def get_hankel(seq_len: int, use_hankel_L: bool = False) -> torch.Tensor:
    """Generate Hankel matrix for spectral filters."""
    entries = torch.arange(1, seq_len + 1, dtype=torch.float64)
    i_plus_j = entries[:, None] + entries[None, :]
    if use_hankel_L:
        sgn = (-1.0) ** (i_plus_j - 2.0) + 1.0
        denom = (i_plus_j + 3.0) * (i_plus_j - 1.0) * (i_plus_j + 1.0)
        Z = sgn * (8.0 / denom)
    else:
        Z = 2.0 / (i_plus_j**3 - i_plus_j)
    return Z

# Note this is slow (usually we just run this once and save the result)
def get_spectral_filters(
    seq_len: int,
    K: int,
    use_hankel_L: bool = False,
    device: torch.device = None,
    dtype: torch.dtype = torch.float32
) -> torch.Tensor:
    """Generate spectral filters using Hankel matrix eigendecomposition."""
    Z = get_hankel(seq_len, use_hankel_L).to(device)
    sigma, phi = torch.linalg.eigh(Z)
    sigma_k, phi_k = sigma[-K:], phi[:, -K:]
    phi_k *= sigma_k ** 0.25
    return phi_k.to(dtype=dtype)

def convolve(
    u: torch.Tensor,
    v: torch.Tensor,
    n: int,
    use_approx: bool = True
) -> tuple[torch.Tensor, torch.Tensor]:
    """Convolve input with filters using FFT."""
    bsz, seq_len, d_in = u.shape
    sgn = torch.full((1, seq_len, 1), 1, device=u.device)
    sgn[:, 1::2] *= -1

    if use_approx:
        _, d_out = v.shape
        v = v.view(1, -1, d_out, 1).to(torch.float32).contiguous()
    else:
        _, K = v.shape
        sgn = sgn.unsqueeze(-1)
        v = v.view(1, -1, K, 1, 1).to(torch.float32).contiguous()

    # Expand u to match v dimensions: [bsz, seq_len, K, d_in]
    u = u.view(bsz, -1, 1, d_in).expand(bsz, -1, K, d_in)
    v = torch.fft.rfft(v, n=n, dim=1)
    U = torch.stack([u, u * sgn], dim=-1).to(torch.float32).contiguous()
    U = torch.fft.rfft(U, n=n, dim=1)
    U_conv = torch.fft.irfft(v * U, n=n, dim=1)[:, :seq_len]
    U_plus, U_minus = torch.unbind(U_conv, dim=-1)
    U_minus = U_minus * sgn
    return U_plus, U_minus

# =============================================================================
# LDS Implementation
# =============================================================================

class LDS(nn.Module):
    """Linear Dynamical System implementation."""
    def __init__(
        self,
        state_dim: int,
        input_dim: int,
        output_dim: int,
        dtype: torch.dtype = torch.float32,
        device: torch.device = None
    ):
        super().__init__()
        self.state_dim = state_dim
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.dtype = dtype
        self.device = device or torch.device('cpu')
        # LDS parameters: h_{t+1} = A * h_t + B * u_t, y_t = C * h_t
        self.A = nn.Parameter(torch.clip(torch.randn(state_dim), -1.0, 1.0).to(dtype).to(self.device))
        self.B = nn.Parameter((torch.randn(input_dim, state_dim) / input_dim).to(dtype).to(self.device))
        self.C = nn.Parameter((torch.randn(state_dim, output_dim) / state_dim).to(dtype).to(self.device))
        self.h0 = nn.Parameter(torch.zeros(state_dim, dtype=dtype).to(self.device))
    
    def forward(self, inputs: torch.Tensor) -> torch.Tensor:
        """Generate LDS trajectory for given inputs."""
        batch_size, seq_len, input_dim = inputs.shape
        outputs = []
        # Initialize hidden state
        h = self.h0.unsqueeze(0).expand(batch_size, -1)
        for t in range(seq_len):
            # Update hidden state: h_{t+1} = A * h_t + B * u_t
            h = h * self.A + inputs[:, t, :] @ self.B
            # Compute output: y_t = C * h_t
            y_t = h @ self.C
            outputs.append(y_t.unsqueeze(1))
        return torch.cat(outputs, dim=1)

def random_LDS(
    state_dim: int,
    input_dim: int,
    output_dim: int,
    lower_bound: float = 0.1,
    device: torch.device = None
) -> LDS:
    """Create a random LDS with specified parameters."""
    device = device or torch.device('cpu')
    lds = LDS(state_dim, input_dim, output_dim, device=device)
    # Custom initialization for A parameter
    A_values = torch.rand(state_dim, device=device) * (1 - lower_bound) + lower_bound
    signs = torch.randint(0, 2, (state_dim,), device=device) * 2 - 1
    lds.A = nn.Parameter((A_values * signs.float()).to(device))
    return lds

# =============================================================================
# Mini STU Implementation
# =============================================================================


class MiniSTU(nn.Module):
    """
    Simplified STU implementation without flash/caching/future-fill.
    Handles batched inputs: x ∈ [B, L, I] -> y ∈ [B, L, O].
    """
    def __init__(
        self,
        seq_len: int,
        num_filters: int,
        input_dim: int,
        output_dim: int,
        use_hankel_L: bool = False,
        dtype: torch.dtype = torch.float32,
        device: torch.device = None,
        default_filters: torch.Tensor = None,  # Optionally provide precomputed filters
    ):
        super().__init__()
        self.seq_len = seq_len
        self.num_filters = num_filters  # K
        self.input_dim = input_dim      # I
        self.output_dim = output_dim    # O
        self.use_hankel_L = use_hankel_L
        self.dtype = dtype
        self.device = device or torch.device("cpu")
        # Spectral filters: shape [L, K]. Register as buffer so moves with .to(device)
        if default_filters is not None:
            phi = default_filters.to(self.device, self.dtype)
        else:
            phi = get_spectral_filters(seq_len, num_filters, use_hankel_L, self.device, self.dtype)
        self.register_buffer("phi", phi, persistent=False)
        # FFT length (if your convolve uses it)
        self.n = nearest_power_of_two(seq_len * 2 - 1, round_up=True)
        # Learnable projections Φ⁺, Φ⁻ : [K, I, O]
        # Small init helps stability when K, I are large
        scale = (num_filters * input_dim) ** -0.5
        self.M_phi_plus = nn.Parameter(
            torch.randn(num_filters, input_dim, output_dim, dtype=dtype, device=self.device) * scale
        )
        if not use_hankel_L:
            self.M_phi_minus = nn.Parameter(
                torch.randn(num_filters, input_dim, output_dim, dtype=dtype, device=self.device) * scale
            )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """
        x: [B, L, I] or [L, I]
        returns: [B, L, O]
        """
        # Allow unbatched input [L, I]
        if x.dim() == 2:
            x = x.unsqueeze(0)  # -> [1, L, I]
        assert x.dim() == 3, f"Expected x with shape [B, L, I] or [L, I]; got {tuple(x.shape)}"
        B, L, I = x.shape
    
        x = x.to(self.M_phi_plus.dtype)
        U_plus, U_minus = convolve(x, self.phi, self.n, use_approx=False)
        # Contract over K and I: [B, L, K, I] ⊗ [K, I, O] -> [B, L, O]
        spectral_plus = torch.einsum('blki,kio->blo', U_plus, self.M_phi_plus)
        if self.use_hankel_L:
            return spectral_plus
        spectral_minus = torch.einsum('blki,kio->blo', U_minus, self.M_phi_minus)
        return spectral_plus + spectral_minus

# =============================================================================
# Training Functions
# =============================================================================

def train_stu_on_lds(
    lds: LDS,
    seq_len: int,
    num_filters: int,
    num_steps: int = 1000,
    batch_size: int = 32,
    learning_rate: float = 1e-3,
    verbose: bool = True
) -> tuple[MiniSTU, list]:
    """Train a STU to learn an LDS."""
    # Create STU model
    stu = MiniSTU(seq_len, num_filters, lds.input_dim, lds.output_dim, device=lds.device, dtype=lds.dtype).to(lds.device)
    # Setup training
    optimizer = torch.optim.Adam(stu.parameters(), lr=learning_rate)
    losses = []
    stu.train()
    for step in tqdm(range(num_steps), desc="Training STU"):
        # Generate random inputs
        inputs = torch.randn(batch_size, seq_len, lds.input_dim, device=lds.device, dtype=lds.dtype)
        # Generate LDS targets (no gradients)
        with torch.no_grad():
            targets = lds(inputs)
        # STU forward pass
        outputs = stu(inputs)
        # Compute loss
        loss = F.mse_loss(outputs, targets)
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        losses.append(loss.item())
        if verbose and step % 100 == 0:
            print(f"Step {step}: Loss = {loss.item():.6f}")
    stu.eval()
    return stu, losses

# =============================================================================
# Visualization and Testing
# =============================================================================

def plot_training_curves(losses: list, title: str = "Training Loss"):
    """Plot training loss curve."""
    plt.figure(figsize=(10, 6))
    plt.plot(losses)
    plt.xlabel('Training Step')
    plt.ylabel('MSE Loss')
    plt.title(title)
    plt.yscale('log')
    plt.grid(True)
    plt.show()

def test_stu_lds_approximation(
    stu: MiniSTU,
    lds: LDS,
    seq_len: int = 1000,
    num_tests: int = 5
):
    """Test how well the STU approximates the LDS."""
    stu.eval()
    lds.eval()
    test_losses = []
    with torch.no_grad():
        for i in range(num_tests):
            # Generate test input
            test_input = torch.randn(1, seq_len, lds.input_dim, device=lds.device, dtype=lds.dtype)
            # Get LDS output
            lds_output = lds(test_input)
            # Get STU output
            stu_output = stu(test_input)
            # Compute test loss
            test_loss = F.mse_loss(stu_output, lds_output)
            test_losses.append(test_loss.item())
            print(f"Test {i+1}: MSE Loss = {test_loss.item():.6f}")
    print(f"Average test loss: {np.mean(test_losses):.6f} ± {np.std(test_losses):.6f}")
    return test_losses

def visualize_predictions(
    stu: MiniSTU,
    lds: LDS,
    seq_len: int = 200,
    input_dim: int = 1,
    output_dim: int = 1
):
    """Visualize STU vs LDS predictions."""
    stu.eval()
    lds.eval()
    with torch.no_grad():
        # Generate test input
        test_input = torch.randn(1, seq_len, input_dim, device=lds.device, dtype=lds.dtype)
        # Get predictions
        lds_output = lds(test_input)
        stu_output = stu(test_input)
        # Plot results
        fig, axes = plt.subplots(2, 1, figsize=(12, 8))
        # Plot input
        axes[0].plot(test_input[0, :, 0].cpu().numpy(), label='Input', alpha=0.7)
        axes[0].set_title('Input Signal')
        axes[0].set_xlabel('Time')
        axes[0].set_ylabel('Value')
        axes[0].grid(True)
        axes[0].legend()
        # Plot outputs
        for i in range(output_dim):
            axes[1].plot(lds_output[0, :, i].cpu().numpy(), label=f'LDS Output {i}', alpha=0.8)
            axes[1].plot(stu_output[0, :, i].cpu().numpy(), '--', label=f'STU Output {i}', alpha=0.8)
        axes[1].set_title('LDS vs STU Predictions')
        axes[1].set_xlabel('Time')
        axes[1].set_ylabel('Value')
        axes[1].grid(True)
        axes[1].legend()
        plt.tight_layout()
        plt.show()

# =============================================================================
# Main Demo
# =============================================================================

def main_demo(
    A_eigenvalues=None,
    seq_len=512,
    num_filters=24,
    num_steps=5000,
    batch_size=16,
    device=None,
    dtype=torch.float32,
):
    """
    Run a complete demo of STU learning LDS, allowing specification of A's eigenvalues.

    Args:
        A_eigenvalues (list or None): List of eigenvalues for the LDS A matrix. If None, defaults to [0.8, 0.9].
        seq_len (int): Sequence length for training and testing.
        num_filters (int): Number of spectral filters in STU.
        num_steps (int): Number of training steps.
        batch_size (int): Training batch size.
        device: Torch device.
        dtype: Torch dtype.

    Returns:
        stu, lds, losses
    """
    print("=== Mini STU Learning LDS Demo ===\n")
    # Set eigenvalues and dimensions
    state_dim = 2 if A_eigenvalues == None else len(A_eigenvalues)
    input_dim = 1
    output_dim = 1

    print(f"State dim: {state_dim}, Input dim: {input_dim}, Output dim: {output_dim}")
    print(f"Sequence length: {seq_len}, Number of filters: {num_filters}")
    print(f"Training steps: {num_steps}, Batch size: {batch_size}\n")

    # Create LDS with specified eigenvalues
    print("Creating LDS with specified eigenvalues...")
    lds = random_LDS(
        state_dim, input_dim, output_dim,
        device=device,
    )
    if A_eigenvalues != None:
        lds.A.data = torch.Tensor(A_eigenvalues)
    print(f"LDS A parameter: {lds.A.data}")
    print(f"LDS B parameter: {lds.B.data}")
    print(f"LDS C parameter: {lds.C.data}\n")

    # Train STU to learn LDS
    print("Training STU to learn LDS...")
    stu, losses = train_stu_on_lds(
        lds, seq_len, num_filters, num_steps, batch_size
    )

    # Plot training curve
    plot_training_curves(losses, "STU Learning LDS - Training Loss")

    # Test approximation quality
    print("\nTesting STU approximation quality...")
    test_losses = test_stu_lds_approximation(stu, lds, seq_len=seq_len)

    # Visualize predictions
    print("\nVisualizing predictions...")
    visualize_predictions(stu, lds, seq_len=seq_len, input_dim=input_dim, output_dim=output_dim)
    return stu, lds, losses

# Run the demo
stu, lds, losses = main_demo()
```

## STU to LDS Distillation

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import sys
import os


class DistillSTU(nn.Module):
    def __init__(self, stu, lds_path=None, state_dim=100, dtype=torch.float32) -> None:
        super(DistillSTU, self).__init__()

        # Extract parameters from MiniSTU (stu)
        # K: number of filters
        self.K = stu.num_filters 
        self.d_in = stu.input_dim
        self.d_out = stu.output_dim 
        self.use_hankel_L = stu.use_hankel_L
        self.use_approx = False #This is about the tensordor approximation
        self.output_dtype = stu.dtype

        #LDS that represents the spectral filters
        self.lds = self.get_lds(lds_path, state_dim, dtype, self.d_in)

        # Copy parameters from MiniSTU
        if self.use_approx:
            raise NotImplemented
        else:
            self.M_phi_plus = nn.Parameter(
                stu.M_phi_plus.data.to(self.lds.dtype)
            )
            if not self.use_hankel_L and hasattr(stu, "M_phi_minus"):
                self.M_phi_minus = nn.Parameter(
                    stu.M_phi_minus.data.to(self.lds.dtype)
                )

    def forward(self, x: torch.Tensor, input_pos: torch.Tensor = None) -> torch.Tensor:
        x = x.to(self.M_phi_plus.dtype)
        bsz = x.shape[0]
        x_reshaped = x.permute(0, 2, 1).reshape(-1, x.shape[1], 1)
        U_reshaped = self.lds(x_reshaped)
        U = U_reshaped.reshape(bsz, x.shape[2], x.shape[1], -1).permute(0, 2, 3, 1)
        U_plus, U_minus = U[:, :, :self.K, :], U[:, :, self.K:, :]
        spectral_plus = torch.tensordot(U_plus, self.M_phi_plus, dims=([2, 3], [0, 1]))
        if not self.use_hankel_L and hasattr(self, "M_phi_minus"):
            spectral_minus = torch.tensordot(U_minus, self.M_phi_minus, dims=([2, 3], [0, 1]))
        ret = spectral_plus if self.use_hankel_L else (spectral_plus + spectral_minus)
        return ret.to(self.output_dtype)

    def loss(self, inputs, targets):
        pred = self.forward(inputs, input_pos=None)
        return F.mse_loss(pred, targets)

    def get_lds(self, checkpoint_path=None, state_dim=100, dtype=torch.float64, bsz_dim=896):
        if checkpoint_path is None:
            print('random init')
            lds_phi = LDS(
                state_dim=state_dim,
                input_dim=1,
                output_dim=2 * self.K,
                dtype=dtype,
                device = device
            )
        else:
            checkpoint = torch.load(checkpoint_path, map_location=torch.device('cpu'))
            lds_phi = LDS(
                state_dim=state_dim,
                input_dim=1,
                output_dim=2 * self.K,
                dtype=dtype,
                device = device
            )
            state_dict = checkpoint['model_state_dict']
            state_dict = {k: v for k, v in state_dict.items() if k not in ['h0', 'M']}
            lds_phi.load_state_dict(state_dict, strict=False)
        return lds_phi.to(device)
```
